# Load modules and Set constants

```python
import os, cv2, glob
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
WIDTH, HEIGHT = 256, 256

EXPERIMENT_DATA = {
    'name' : 'Trajectory_Clusetring',
    'number' : '31',
    'date' : '07-15-2021',
    'description' : 'Check Low MSE Images\n'
    }

ROOT_DIR = os.path.join('C:\\', 'GitHub', 'Lab_Experiment', 'Trajectory_Clustering')

GEOLIFE_DIR = os.path.join('C:\\', 'Dataset', 'Geolife')
IMAGE_DIR = os.path.join(GEOLIFE_DIR, 'Image_Files')

RESULT_DIR = os.path.join( 'C:\\', 'GitHub', 'Lab_Experiment', 'Results')

os.chdir(RESULT_DIR)
curr_result_name = f"{EXPERIMENT_DATA['name']}_{EXPERIMENT_DATA['number']}_{EXPERIMENT_DATA['date']}_Results"
if (curr_result_name not in os.listdir()):
    os.mkdir(curr_result_name)

CURR_RESULT_DIR = os.path.join(RESULT_DIR, curr_result_name)
os.chdir(CURR_RESULT_DIR)

with open('Description.txt', 'w') as f:
    f.write(EXPERIMENT_DATA['description'])

print(f'ROOT_DIR:          {ROOT_DIR}')
print(f'GEOLIFE_DIR:       {GEOLIFE_DIR}')
print(f'IMAGE_DIR:         {IMAGE_DIR}')
print(f'RESULT_DIR:        {RESULT_DIR}')
print(f'CURR_RESULT_DIR:   {CURR_RESULT_DIR}')
```

# Check Low MSE Images

```python
os.chdir(GEOLIFE_DIR)
map_table = pd.read_csv('Map_Table.csv', header = 0)

os.chdir(IMAGE_DIR)
plt.gray()
plt.imshow(cv2.imread('Geolife_trajectory1.png', cv2.IMREAD_GRAYSCALE))
plt.show()
```

## Check 0.01 threshold MSE Images

```python
os.chdir(CURR_RESULT_DIR)
df = pd.read_csv('MSE_RESULT_0.01.csv', header = 0)

df = df.sort_values(by = 'MSE', axis = 0)
```

```python
df.head()
```

```python
os.chdir(IMAGE_DIR)

n = 10
fig, ax = plt.subplots(1, n, sharex = True, sharey = True, figsize = (80, 8))
for i in range(n):
    image_file_name = map_table.loc[map_table[' LowDimNumber'] == int(df['DataNumber'].iloc[i].split('th')[0])]
#     print(image_file_name)
    ax[i].imshow(cv2.imread(f'Geolife_trajectory{image_file_name.iloc[0]["ImageNumber"]}.png', cv2.IMREAD_GRAYSCALE))
    ax[i].set_title(f'MSE: {df["MSE"].iloc[i]}', fontsize = 20)

plt.suptitle(f'Threshold: {0.01}', fontsize = 30)

os.chdir(CURR_RESULT_DIR)
plt.savefig(f'Result({0.01}).png', dpi = 100)
plt.show()
```
